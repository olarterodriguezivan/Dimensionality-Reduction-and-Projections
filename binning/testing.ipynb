```python
from binning_heuristics import BinningHeuristic
```

```python
import numpy as np

# Create some sample data
rng = np.random.default_rng(42)

y = rng.normal(loc=0.0, scale=1.0, size=1000)

# Initialize the BinningHeuristic with a specific method
binning = BinningHeuristic(method='sturges')
binning.fit(y)
transformed_y = binning.transform(y, zero_index=True)

print("Bin edges:", binning.bin_edges)
print("Number of bins:", binning.n_bins)
print("Variable range:", binning.variable_range)
```
